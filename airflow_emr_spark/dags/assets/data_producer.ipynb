```python
import os
import pandas as pd
import numpy as np
import boto3
import awswrangler as wr

bucket = 'xxxxxx'
object_key = 'src/balances'

boto3.setup_default_session(aws_access_key_id = 'xxxxxx',
                            aws_secret_access_key = 'xxxxxx')


# Generate mock data
num_rows = 10000000 
data = {
    'AS_OF_DATE': np.random.choice(pd.date_range(start='2023-01-01', end='2023-08-31'), size=num_rows),
    'COMPANY_CODE': np.random.choice(['Company_A', 'Company_B', 'Company_C', 'Company_D'], size=num_rows),
    'ACCOUNT_NAME': np.random.choice(['Assets', 'Liabilities', 'Cogs', 'Inter Company'], size=num_rows),
    'ACCOUNT_NUMBER': np.random.randint(100000, 999999, size=num_rows).astype(str),
    'BALANCE': np.random.uniform(low=0, high=500000, size=num_rows).astype(np.float32),
    'CURRENCY_CODE': np.random.choice(['USD', 'EUR', 'GBP'], size=num_rows),
}

# Create the dataframe
df = pd.DataFrame(data).sort_values(by=['COMPANY_CODE', 'AS_OF_DATE'])

print(df.shape)
df.head(10)

# Write DF to S3 Bucket 
wr.s3.to_parquet(df=df, path=f's3://{bucket}/{object_key}', compression = 'snappy')
```
